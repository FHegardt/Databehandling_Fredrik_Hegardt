## Setup

```python
#create dummy data

import pandas as pd 
import numpy as np

np.random.seed(42)
n = 10
random_matrix = np.random.randint(1,10, (n,n)) #2D array
index = np.random.choice(random_matrix.size, 10, replace=False)
print(index)
random_matrix = random_matrix*1.0
random_matrix.ravel()[index] = None
```

```python
scores = pd.DataFrame(
            random_matrix, index =[f"Player {i}" for i in range(1, n+1)],
            columns= [f"Round {i}" for i in range(1, n+1)])
scores.head()
```

```python
scores.info()
```

## Missing value strategy

Strategy depends on

- Dataset Size
- Valueable info (some rows, some columns)
- percentage missing values
- domain knowledge 
 - data visualization
 - arithmetic computation
 - summary statistics
 - machine learning algoritms

```python
import seaborn as sns

titanic =sns.load_dataset("titanic")
titanic.head()
```

```python
titanic.isnull().sum()
```

```python
#focus on age
sns.set_theme()
sns.histplot(data=titanic,x="age", kde=True, hue="sex", multiple = "dodge")
```

```python
# talked to a titanic historical
# this dude say: use median age of corresponding gender to fill in ages

print(f"Number of mens: {sum(titanic['sex'] == 'male')}")
print(f"Number of females: {sum(titanic['sex'] == 'female')}")

median_male_age = titanic.loc[titanic['sex'] == 'male', "age"].median()
median_female_age = titanic.loc[titanic['sex'] == 'female', "age"].median()

median_male_age, median_female_age
```

```python
titanic.loc[titanic["sex"]== "male", "age"] = titanic.loc[titanic["sex"]== "male", "age"].fillna(median_male_age)
titanic.loc[titanic["age"]==29, "sex"].value_counts()
```

```python
titanic.isnull().sum()["age"]
titanic.loc[titanic["sex"]== "female", "age"] = titanic.loc[titanic["sex"]== "female", "age"].fillna(median_female_age)
```
