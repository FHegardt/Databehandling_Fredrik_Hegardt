```python
import random as rnd

def System1(low, high):
    return rnd.randint(low,high)
def System2(low, high):
    return int((high+low)/2)

def my_number(low, high):
    return rnd.randint(low,high)

def guessing_game(guess, low, high, system):
    current_number = my_number(low, high)
    guess_nr = 0
    while current_number != guess:
        guess_nr += 1
        guess = system(low, high)
        print(f"System 1: My {guess_nr} guess is {guess}")
        if guess == current_number:
            print("AnswerSystem: That is correct!")
        elif guess > current_number:
            high = guess
            print("AnswerSystem: Wrong! Lower...")
        elif guess < current_number:
            low = guess
            print("AnswerSystem: Wrong! Higher...")
    return guess_nr
            

```

```python
list1 = []
list2 = []
for i in range(10):
    list1.append(guessing_game(0,1,100, System1))
    list2.append(guessing_game(0,1,100, System2))
print(list1)
print(list2)
```
