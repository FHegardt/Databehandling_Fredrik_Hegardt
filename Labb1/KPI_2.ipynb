## KPI2 = Antal intensivvårdade per dag vs antal avlidna per dag

```python
import pandas as pd
import seaborn as sns
import numpy as np

smittodata = pd.ExcelFile("Folkhalsomyndigheten_Covid19.xlsx")
av_per_dag = pd.read_excel(smittodata, 'Antal avlidna per dag')
iva_per_dag = pd.read_excel(smittodata, 'Antal intensivvårdade per dag')
per_ålder = pd.read_excel(smittodata, 'Totalt antal per åldersgrupp')
per_region = pd.read_excel(smittodata, 'Totalt antal per region')
iva_per_dag.head()
```

```python
import plotly_express as px

fig1 = px.bar(av_per_dag, x= "Datum_avliden", y = "Antal_avlidna", title= "Antal Covid-19 avlidna vs intensivvårdade per dag")
fig1.add_scatter(x=iva_per_dag["Datum_vårdstart"],y=iva_per_dag["Antal_intensivvårdade"])
fig1.show()

fig1.write_html("Visualiseringar/KPI2.html")
```

## KPI3 = Antal intensivvårdade vs antal avlidna per dag och åldersgrupp

```python
fig1 = px.bar(per_ålder, x= "Åldersgrupp", y = "Totalt_antal_avlidna", title= "Avlidna vs intensivvårdade per åldergrupp")
fig1.add_bar(x=per_ålder["Åldersgrupp"],y=per_ålder["Totalt_antal_intensivvårdade"])
fig1.show()
```
